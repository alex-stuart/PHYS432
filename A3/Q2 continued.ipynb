```python
#@author: Alexandre Stuart. Collaborators: Ruijia Yang and Yacine Benkirane 

import numpy as np
import matplotlib.pyplot as plt
from sympy.solvers import solve
from sympy import Symbol
#%matplotlib inline

x = Symbol('x', real=True)

#making an array of equally spaced nu values for the x axis
nu_vals = np.logspace(-2,2,100)
rho=1
r=10
h=200
#finding the velocities
v = [solve(np.pi*rho*x**2*r*(3*nu+2*h*x*10**(-3)+x*r)-3.27*10**8, x) for nu in nu_vals]
plt.xscale('log')

#substances with different viscosities
plt.plot([0.01,0.01],[91,100],'b') # Water
plt.plot([0.4,0.4],[91,100],'g') # Canola Oil
plt.plot([16.50,16.50],[91,100],'r') # Glucose
plt.plot([0.064,0.064],[91,100],'k') # Gasoline
plt.plot([30,30],[91,100],'m') #Molasses 
plt.plot(nu_vals,v)
plt.title('Swimmer speed as function of fluid viscosity')
plt.legend(['Water', 'Canola Oil','Glucose','Gasoline','Molasses'])
plt.xlabel('Fluid viscosity (cm$^2$/s)')
plt.ylabel('Swimmer speed (cm/s)')
plt.show
```

All of the viscosity values were taken from https://www.engineeringtoolbox.com/kinematic-viscosity-d_397.html 

This result seems to make sense as intuitively, swimming speed is greater in less viscous fluids
